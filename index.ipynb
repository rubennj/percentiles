```python
import pandas as pd
import matplotlib.pyplot as plt

fig, (ax_peso, ax_talla) = plt.subplots(nrows=2, ncols=1, figsize=(12, 8), sharex=True)

# peso
percentiles_peso = pd.read_csv('wtageinf.csv', index_col='Agemos')

pesadas = pd.read_csv('pesadas.txt', index_col='fecha', parse_dates=True, dayfirst=True)
pesadas['meses'] = (pesadas.index - pesadas.index[0]).days / 30

percentiles_peso.query('Sex==1')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(ax=ax_peso, cmap='jet', legend=None, linewidth=5, alpha=0.5)
pesadas.plot.scatter(ax=ax_peso, x='meses', y='peso', color='black', marker='x')

ax_peso.legend(ncol=3, loc='lower right', fontsize=10, framealpha=0.15)

ax_peso.set_xlim([0, 6])
ax_peso.set_ylim([2, 9])

ax_peso.grid(True)

# talla
percentiles_talla = pd.read_csv('lenageinf.csv', index_col='Agemos', skiprows=[38])

talladas = pd.read_csv('talladas.txt', index_col='fecha', parse_dates=True, dayfirst=True)
talladas['meses'] = (talladas.index - talladas.index[0]).days / 30

percentiles_talla.query('Sex==1')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(ax=ax_talla, cmap='jet', legend=None, linewidth=5, alpha=0.5)
talladas.plot.scatter(ax=ax_talla, x='meses', y='talla', color='black', marker='x')

ax_talla.legend(ncol=3, loc='lower right', fontsize=10, framealpha=0.15)

ax_talla.set_xlim([0, 6])
ax_talla.set_ylim([45, 70])

ax_talla.grid(True)
```
